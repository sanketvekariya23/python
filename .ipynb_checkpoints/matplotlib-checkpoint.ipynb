```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x = [4,5,6]
y = [4.5,6,7]
x1= [10,8,4]
y1 = [3,2,1]

plt.plot(x,y,label = "first",linewidth = 4,linestyle = 'dotted')
plt.plot(x1,y1, label = "second")
plt.legend(loc = 'best',shadow = True)
plt.show()
```

# bar chart

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
compny = ["google" ," facebook" ," amzone" , "flpcart"]
revenue= [96,129,87,56]
profit = [56,58,60,50]
```

```python
xpos = np.arange(len(compny))
xpos
```

```python
plt.bar(xpos , revenue, label='revenue')
plt.bar(xpos,profit ,label="profit")
plt.legend()
plt.ylabel("revenue")
plt.xlabel("compny")
plt.title("compney charts")
plt.xticks(xpos,compny)
```

# horizotal bar charts


```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

compny = ["google" ," facebook" ," amzone" , "flpcart"]
revenue= [96,129,87,56]
profit = [56,58,60,50]

xpos = np.arange(len(compny))

plt.barh(xpos , revenue, label='revenue')
plt.barh(xpos,profit ,label="profit")
plt.legend(Shadow = True)
plt.ylabel("revenue")
plt.xlabel("compny")
plt.title("compney charts")
plt.yticks(xpos,compny)
```
