```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x = [4,5,6]
y = [4.5,6,7]
x1= [10,8,4]
y1 = [3,2,1]

plt.plot(x,y,label = "first")
plt.plot(x1,y1, label = "second")
plt.legend()
plt.show()
```
